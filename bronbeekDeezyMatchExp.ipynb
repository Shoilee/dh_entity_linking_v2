```python
def experiment200():
    dm_inference(os.path.join("inputs", "input_dfm.yaml"),
                 dataset_path=os.path.join("bronbeek", "queries.txt"), 
                 pretrained_model_path=os.path.join("models", "test001", "test001.model"), 
                 pretrained_vocab_path=os.path.join("models", "test002", "test002.vocab"),
                 inference_mode="vect",
                 scenario="queries/exp200")
    
    # generate vectors for candidates (specified in dataset_path) 
    # using a model stored at pretrained_model_path and pretrained_vocab_path 
    dm_inference(os.path.join("inputs", "input_dfm.yaml"),
                 dataset_path=os.path.join("bronbeek", "candidates.txt"), 
                 pretrained_model_path=os.path.join("models", "test001", "test001.model"), 
                 pretrained_vocab_path=os.path.join("models", "test001", "test001.vocab"),
                 inference_mode="vect",
                 scenario="candidates/exp200")
    
    # combine vectors stored in queries/test and save them in combined/queries_test
    combine_vecs(rnn_passes=['fwd', 'bwd'], 
                 input_scenario=os.path.join('queries', 'exp200'), 
                 output_scenario=os.path.join('combined', 'queries_exp200'), 
                 print_every=10)
    
    combine_vecs(rnn_passes=['fwd', 'bwd'], 
                 input_scenario=os.path.join('candidates', 'exp200'), 
                 output_scenario=os.path.join('combined', 'candidates_exp200'), 
                 print_every=10)
    
    # Select candidates based on L2-norm (aka faiss distance):
    # find candidates from candidate_scenario 
    # for queries specified in query_scenario
    candidates_pd = \
                candidate_ranker(query_scenario=os.path.join("combined", "queries_exp200"),
                    candidate_scenario=os.path.join("combined", "candidates_exp200"), 
                    ranking_metric="faiss", # two accepted value = ['cosine', faise]
                    selection_threshold=.5, 
                    num_candidates=3, 
                    search_size=10, 
                    verbose=False,
                    use_predict=False,
                    output_path=os.path.join("ranker_results", "exp200_candidates_deezymatch"), 
                    pretrained_model_path=os.path.join("models", "test001", "test001.model"), 
                    pretrained_vocab_path=os.path.join("models", "test001", "test001.vocab"))
    # TODO print "ranker_results/exp200_candidates_deezymatch"
```
